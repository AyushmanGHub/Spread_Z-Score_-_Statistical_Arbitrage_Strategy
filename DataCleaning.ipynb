```python
# loading needed libraries

import pandas as pd
import os

data_dir = 'data'
```

```python
data_path = os.path.join(data_dir, "data.parquet")
df = pd.read_parquet(data_path)
df.head(5)
```

### Ensuring datetime index

```python
if "time" in df.columns:  
    df["time"] = pd.to_datetime(df["time"])
    df = df.set_index("time").sort_index()
else:
    df.index = pd.to_datetime(df.index)
    df = df.sort_index()
```

```python
print("Missing values BEFORE cleaning")
print(df.isna().sum())
print(f"Total rows before cleaning: {len(df)}\n")
```

```python
# Dropping all rows with ANY NaN
df_cleaned = df.dropna()
```

```python
print("Missing values AFTER cleaning")
print(df_cleaned.isna().sum())
print(f"Total rows after cleaning: {len(df_cleaned)}")
print(f"Dropped {len(df) - len(df_cleaned)} rows\n")
```

```python
processed_data_path = os.path.join( data_dir, "processed_data.parquet")
df_cleaned.to_parquet(processed_data_path, engine="pyarrow", index=True)
```
